```python
# read the data from the file and handle null values

import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(20)
```

```python
# rename the columns

df_col=df.columns.str.lower()
df.columns=df_col.str.replace(' ','_')
df.head(20)
```

```python
# create columns - discount,sale price,profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['discount']=df['discount'].round(3)
df.head()
```

```python
df['sale_price']=df['list_price']-df['discount']
df.head(20)
```

```python
df['profit']=df['sale_price']-df['cost_price']
df['profit']=df['profit'].round(3)
df.head(20)
```

```python
# change the date format from 'object' to 'date'
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```

```python
# load the data into SQL server

import os
import pyodbc
from sqlalchemy import create_engine
```

```python
print(pyodbc.drivers())
```

```python
server_name='IamSKS'
database_name='orders'
driver = 'ODBC Driver 17 for SQL Server'
database_uri=f'mssql+pyodbc://@{server_name}/{database_name}?driver={driver}&trusted_connection=yes'
engine=create_engine(database_uri)

df.to_sql('orders',con=engine,if_exists='append',index=False)
```
